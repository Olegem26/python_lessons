```python
'''LEGB (области вимости в коде) - это правило, которое определяет порядок поиска переменной в коде.'''
```

```python
#Аббревиатура расшифровывается как Local(L), Enclosing(E), Global(G), Built-in(B).
```

```python
# Local - область видимости переменной внутри одной определенной функции.
def number():
    count = 0 #local
    print(count)
    return number

number()
```

```python
# Enclosing - область видимости переменной во внешней функции (охватывающей).
def number():
    count = 0 # Enclosing для функции func
    
    def func():
        value = 23 # Enclosing для функции in_fun? которая находится в func
        print(count)

        def in_fun():
            print(value)
        in_fun()
    
    func()

    return number

number()
    
```

```python
# Global - область видимости переменной во всем коде, в лубой части программы
count = 5

def func():
    print(count)
func()
```

```python
# Built-in - специальная область видимости, которая содержит встроенные функции, типы, исключения и другие атрибуты из питона.
print('Hello')# print - это функция из области видимости Built-in.
```

```python
""" Как работает это в Python?
Когда интерпритатор встречает имя переменной, то он начинает искать её сначала в локальной области видимости, затем в охватывающей,
потом в глобальной, и в конце во вcтроенной. Если не найдено ни в одной из области, то выведет ошибку."""
```
