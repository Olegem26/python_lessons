```python
#set - что это такое:
''' set( множество ) - это неупорядоченая коллекция из уникальных элементов, в котором автоматически удаляются
 повторяющиеся значения '''
#создадим множество:
new_set = {1, 2, 2, 3, 4, 5} #такое множество сразу удалит повторную цифру 2
new_set
# также множества можно создавать с помощью функции set()
```

```python
new_list = [1, 2, 2, 3]
new_set = set(new_list)
new_set
```

```python
new_set.add(3)
new_set.add(4)
new_set
```

```python
#tuple - что это такое:
''' tuple( кортеж ) - это упорядоченая неизменяемая коллекция элементов'''
# создадим кортеж, есть несколько способов:
#1. через круглые скобочки:
new_tuple = (1, 2, 3)
#2. без скобок
new__nuple = 1, 2, 3
#3. с помощью конструктора tuple()
empty_tuple = tuple()  # пустой кортеж
from_string = tuple("abc")  # ('a', 'b', 'c')



```

```python
def fun():
    return 1, 2

res = fun()
type(res) 
```

```python
# В чем их отличие от list( списков )
""" кортежи и списки:
списки можно изменять, а кортежи нет.
для списков есть много методов (append, remove, sort и т.д), а для кортежей всего два( count, index).
кортеж занимает меньше памяти, чем список( сто и объясняется большим количеством методов в списках) 
сл-но скорость обработки у кортежа быстрее
"""
```

```python
import sys

l1 = [1, 2, 3, 4, 5]
t1 = 1, 2,3,4,5

print(sys.getsizeof(l1))
print(sys.getsizeof(t1))
```
