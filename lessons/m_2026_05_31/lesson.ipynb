```python
# Для добавления елемента в конец списка сложность О(1)
number_elements = 10_000_000
l1 = []
for i in range(number_elements):
    l1.append(i)
len(l1)
```

```python
# Для добавления елемента в начало списка сложность О(N)
number_elements = 100_000
l1 = []
for i in range(number_elements):
    l1.insert(0, i)
print(len(l1))
l1
```
